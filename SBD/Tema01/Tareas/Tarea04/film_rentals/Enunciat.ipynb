### 2.3. Amplia el catàleg de pel·lícules.

Cada tenda té 4 còpies de cada `film_title`. El propietari de l'empresa vol saber quines películes de les que té l'interesa comprar-ne còpies. 

Dona una llista de 10 pel·lícules per tenda i explica quins criteris seguiràs per crear aquestes llistes. 

Pots ajudar-te dels dos nous DataFrames: `renting_hours_store1` i `renting_hours_store2`. Cada fila et diu el títol, el nombre de lloguers d'aquesta cinta simultanis i quantes hores s'ha donat aquesta situació. Per exemple: `ACADEMY DINOSAUR` ha passat `23.963889` hores on hi havia les 4 còpies llogades a l'hora.

```python
# Funció per calcular temps total amb n lloguers simultanis per film
def rental_overlap_durations(sub_df):
    # Crear esdeveniments
    events_start = sub_df[["rental_date"]].copy()
    events_start["delta"] = 1
    events_end = sub_df[["return_date"]].copy()
    events_end["delta"] = -1

    events_start = events_start.rename(columns={"rental_date": "timestamp"})
    events_end = events_end.rename(columns={"return_date": "timestamp"})
    events = pd.concat([events_start, events_end], ignore_index=True)

    # Ordenar per temps i actualitzar nombre actiu
    events = events.sort_values(by=["timestamp", "delta"])
    events["active"] = events["delta"].cumsum()

    # Calcular temps entre events consecutius
    events["next_time"] = events["timestamp"].shift(-1)
    events["duration"] = (events["next_time"] - events["timestamp"]).dt.total_seconds() / 3600  # hores

    # Ignorar última fila (no té següent)
    events = events.dropna(subset=["duration"])

    # Sumar temps per cada nombre de lloguers actius
    result = (
        events.groupby("active")["duration"]
        .sum()
        .reset_index()
        .rename(columns={"active": "simultaneous_rentals", "duration": "hours"})
    )

    return result


# Aplicar per cada film de cada botiga
renting_hours_store1 = (
    df[df["store_id"] == 1]
      .groupby("film_title")
      .apply(rental_overlap_durations, include_groups=False)
      .reset_index(level=0)
      .reset_index(drop=True)
)

renting_hours_store2 = (
    df[df["store_id"] == 2]
      .groupby("film_title")
      .apply(rental_overlap_durations, include_groups=False)
      .reset_index(level=0)
      .reset_index(drop=True)
)
```
