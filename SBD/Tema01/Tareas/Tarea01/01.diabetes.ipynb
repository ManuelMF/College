```python
# 4 Crear una nova sèrie “Quality-color”.
def ratioPregnaciesColor(price):
    if price > 600:
        return "red"
    elif price > 500:
        return "yellow"
    else:
        return "green"
    
df['Quality-color'] = df['price'].apply(ratioPregnaciesColor)

print(df)
```
