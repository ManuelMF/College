2. El Enfoque Orientado a Objetos

Aunque pyplot es excelente para tareas rápidas, el enfoque orientado a objetos nos proporciona un control absoluto sobre cada elemento de una figura, siendo esencial para visualizaciones complejas y de calidad de publicación.

Antes de empezar, es clave entender la anatomía de un gráfico de Matplotlib. Imagina una jerarquía de objetos:

Figure

Es todo el lienzo, la ventana principal que contiene la visualización completa. Puedes imaginarlo como el marco de un cuadro.

Axes

Son los gráficos individuales dentro de la figura. Un Axes contiene los ejes X e Y, las marcas, las etiquetas y las propias curvas o datos representados.
Una sola figura puede contener múltiples Axes (subgráficos).

Elementos menores

Dentro de cada Axes hay componentes más pequeños como:

Spines (las líneas del borde del gráfico),

Ticks (las marcas de los ejes),

Labels (títulos y etiquetas).

Estos ejercicios están diseñados para desarrollar vuestra habilidad en la manipulación directa de los objetos Figure y Axes.

Ejercicio 4: Recrear un Gráfico con subplots()
Objetivo

Adaptar el código de un gráfico simple al enfoque orientado a objetos.
Este ejercicio refuerza la comprensión de que los objetos Figure y Axes son los contenedores de las visualizaciones.

Datos

Usa los arrays de NumPy definidos en la preparación:

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(x, df.facecream,    color='r',      label='Vendes de Face Cream')
ax.plot(x, df.facewash,     color='b',      label='Vendes de Face Wash')
ax.plot(x, df.toothpaste,   color='y',      label='Vendes de Tooth Paste')
ax.plot(x, df.bathingsoap,  color='brown',  label='Vendes de Bathing Soap')
ax.plot(x, df.shampoo,      color='orange', label='Vendes de Shampoo')
ax.plot(x, df.moisturizer,  color='black',  label='Vendes de Moisturizer')

ax.set_xlabel("Número de mes")
ax.set_ylabel("Vendes totals")
ax.set_title("Vendes mensuals per producte")

ax.legend()

plt.show()
```

Pista: Les coordenades a fig.add_axes([esquerra, baix, amplada, alçada]) són relatives a la mida de la figura, en un rang de 0 a 1. Experimenta canviant aquests valors per veure com es mou i canvia de mida el gràfic inserit.

Havent dominat el control sobre la disposició dels gràfics, ja esteu preparats per explorar la diversitat de tipus de gràfics que Matplotlib ofereix per representar diferents relacions entre dades.



```python
fig = plt.figure()

x = np.linspace(0, 5, 11)
y = x**2

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y)

ax.set_title("Grafica")
ax.set_xlabel("X")
ax.set_ylabel("Y")


ax_inserit = fig.add_axes([0.2, 0.5, 0.4, 0.3])

ax_inserit.plot(x**2, x)

ax_inserit.set_title("Grafica")
ax_inserit.set_xlabel("X")
ax_inserit.set_ylabel("Y")
```
