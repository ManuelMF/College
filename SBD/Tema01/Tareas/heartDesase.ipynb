### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'></span> DESCRIPCIÓ DE LES VARIABLES </b>

<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
El conjunt de dades consta de la següent informació de 303 pacients:
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>age</code> : Edat </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
L'edat del pacient en anys. Representada per un nombre enter.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>sex</code> : Sexe </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Sexe del pacient. Representada per un indicador binari, <code>0</code> correspon al sexe femení i <code>1</code> al sexe masculí.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>cp</code> : Categoria del dolor toràcic </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Categorització del dolor toràcic. Cada categoria està representada per un valor enter: <code>1</code> angina típica; <code>2</code> angina atípica; <code>3</code> dolor no anginós; <code>4</code> asimptomàtic.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>trestbps</code> : Pressió arterial en repòs </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Mesura de la pressió arterial en repòs, amb unitats de mm Hg, del pacient en el moment d'ingrés hospitalari representat per un enter.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>chol</code> : Colesterol sèric </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Mesura del colesterol sèric del pacient en mg/dl.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>fbs</code> : Glucosa en dejú </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Mesura de la glucosa en sang en dejú. Variable binària: <code>1</code> si és &gt; 120 mg/dl, <code>0</code> en cas contrari.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>restecg</code> : Electrocardiograma en repòs </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Resultats de l'ECG en repòs: <code>0</code> normal, <code>1</code> anomalies de l'ona ST-T, <code>2</code> hipertrofia ventricular esquerra segons els criteris d'Estes.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>thalach</code> : Freqüència cardíaca màxima </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Freqüència cardíaca màxima assolida pel pacient durant una prova d'esforç, mesurada en bpm.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>exang</code> : Angina induïda per l'exercici </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Indica si el pacient presenta angina induïda durant la prova d'esforç: <code>1</code> sí, <code>0</code> no.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>oldpeak</code> : Depressió ST </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Mesura la quantitat de mil·lívolts (<i>mV</i>) en què el segment ST de l’electrocardiograma queda deprimit respecte al nivell basal durant la prova d’esforç.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>slope</code> : Pendent del segment ST </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Codifica la forma del segment ST de l’electrocardiograma durant la prova d’esforç. S’interpreta de la manera següent: <code>1</code> ascendent (normal); <code>2</code> pla (pot indicar isquèmia); <code>3</code> descendent (més sospitós d’isquèmia).
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>thal</code> : Resultats del test nuclear d'estrès </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Resultat del test nuclear amb tal·li: <code>3</code> normal, <code>6</code> defecte fix, <code>7</code> defecte reversible.
</p>

#### <b style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; color:#00000;"> <span style='color:#009FE3;'>✦ </span><code>num</code> : Diagnòstic de malaltia coronària </b>
<p style="font-family:'Segoe UI', 'Lucida Grande', Verdana, sans-serif; font-size:15px; line-height:1.6; color:#00000;">
Variable objectiu que representa el nombre de vasos coronaris principals amb estenosi significativa (&gt;50%). Pot prendre valors de <code>0</code> a <code>4</code>, on: <code>0</code> és cap vas afectat (absència de malaltia coronària); <code>1</code> a <code>3</code>és el nombre de vasos amb estenosi significativa; <code>4</code> correspon a malaltia del tronc principal esquerre (afectació greu).
</p>
