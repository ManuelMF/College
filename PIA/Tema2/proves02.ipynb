```python
from sklearn.preprocessing import StandardScaler
```

* StandardScaler() (default) -> z = x - mu / sigma. K Nearest Neighbors
* RobustScaler() -> Standard pero robusto a outliers
* MinMaxScaler() -> [0, 1] Multinomial Naïve Bayes
* No escalar -> Decision Tree

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
#cylinders i model_year
#Ordinal puc deixar les categories en 'auto' ja que 
#els valors s'ordenaran alfabèticament i és correcte
#sino categories=['primer valor', 'segon valor', 'tercer valor'...]
enc = OrdinalEncoder()
ordinal_transformed_array = enc.fit_transform(mpg_df[['cylinders', 'model_year']])
mpg_transformed = pd.DataFrame(data=ordinal_transformed_array, columns=['cylinders', 'model_year'])
mpg_transformed
```

```python
from sklearn.preprocessing import OneHotEncoder
#nominal -> origin, name
enc = OneHotEncoder(handle_unknown='error')
nominals_transformades = enc.fit_transform(mpg_df[['origin', 'name']]).toarray()
noms_columnes = enc.get_feature_names_out(['origin', 'name'])
nominals_df = pd.DataFrame(data=nominals_transformades, columns=noms_columnes)
mpg_transformed = pd.concat([mpg_transformed, nominals_df], axis=1) #concatenam columnes
mpg_transformed
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
mpg_scaled = scaler.fit_transform(mpg_df[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']])
mpg_transformed[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']] = mpg_scaled
mpg_transformed
```

```python
#Característique onehot millor al final
columnes = mpg_transformed.columns.tolist()
columnes = columnes[-5:] + columnes[:-5]
mpg_transformed = mpg_transformed[columnes]
mpg_transformed
```
