```python
ratings = pd.read_csv('ml-25m/ratings.csv')
ratings.head()
```

```python
movies = pd.read_csv('ml-25m/movies.csv')
movies.set_index('movieId',inplace=True)
movies.head()
```

```python
ratings_grouped = ratings.groupby('movieId')
ratings_grouped
```

```python
movies['ratingCount'] = ratings_grouped['userId'].count()
movies
```

```python
popular_movies = movies[movies['ratingCount'] > 1000]
popular_movies
```

```python
movie_rating = popular_movies.merge(ratings, left_on='movieId', right_on='movieId')
movie_rating
```

```python
movie_rating.drop(columns=['movieId', 'genres', 'ratingCount', 'timestamp'], inplace=True)
```

```python
movie_rating
```

```python
moviemat = pd.pivot_table(movie_rating, index='userId', columns='title', values='rating')
moviemat
```

```python
moviemat.to_csv('moviemat.csv')
```

```python
terminator_ratings = moviemat['Terminator, The (1984)']
terminator_ratings
```

```python
similar_to_terminator = moviemat.corrwith(terminator_ratings)
similar_to_terminator = pd.DataFrame(similar_to_terminator, columns=['correlacio'])
similar_to_terminator.dropna(inplace=True)
```

```python
similar_to_terminator.sort_values('correlacio', ascending=False).head(10)
```
